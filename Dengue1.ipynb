```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
dengue_data = pd.read_csv("Dengue set.csv")
```

```python
dengue_data.head()
```

```python
dengue_data.info()
```

```python
dengue_data.isnull().sum()
```

```python
dengue_data['Temperature'] = dengue_data['Temperature'].astype(float)
```

```python
data = dengue_data['Temperature']
ks_stat, ks_pvalue = stats.kstest(data, 'norm')
print('Kolmogorov-Smirnov Test statistic:', ks_stat, 'P-value:', ks_pvalue)

plt.hist(data, bins=50, alpha=0.75, color='blue')
plt.title('Histogram of Data')
plt.show()

# Q-Q Plot
stats.probplot(data, dist="norm", plot=plt)
plt.title('Q-Q Plot')
plt.show()
```

```python
data = pd.DataFrame({
    'var1': dengue_data['Temperature'],
    'var2': dengue_data['Dew Point©']
})


# Calculate Spearman's Rank Correlation
spearman_corr, spearman_pval = stats.spearmanr(data['var1'], data['var2'])
print("Spearman's rank correlation coefficient:", spearman_corr)
print(f"P-value: {spearman_pval:.10e}")
```
