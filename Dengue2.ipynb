```python

import pandas as pd


data = pd.read_csv("Dengue.csv")

# Step 1: Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Step 2: Convert the 'Data' column to numeric, forcing non-numeric values to NaN
data['Data'] = pd.to_numeric(data['Data'], errors='coerce')

# Step 3: Create a complete date range
start_date = data['Date'].min()
end_date = data['Date'].max()
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# Step 4: Reindex the original data to the complete date range
daily_data = data.set_index('Date').reindex(date_range).rename_axis('Date').reset_index()

# Step 5: Fill the missing 'Data' values using forward fill method
daily_data['Data'] = daily_data['Data'].fillna(method='ffill')

# Step 6: Evenly distribute the weekly values over 7 days
daily_data['Data'] = daily_data['Data'] / 7

# Display the first few rows of the daily data
print(daily_data.head(20))
```

```python
import pandas as pd


data = pd.read_csv("Dengue set1.csv")

data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Step 2: Convert the 'Data' column to numeric, forcing non-numeric values to NaN
data['Dengue Data'] = pd.to_numeric(data['Dengue Data'], errors='coerce')

# Step 3: Create a complete date range
start_date = data['Date'].min()
end_date = data['Date'].max()
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# Step 4: Reindex the original data to the complete date range
daily_data = data.set_index('Date').reindex(date_range).rename_axis('Date').reset_index()

# Step 5: Fill the missing 'Data' values using forward fill method
daily_data['Dengue Data'] = daily_data['Dengue Data'].fillna(method='ffill')

# Step 6: Evenly distribute the weekly values over 7 days
daily_data['Dengue Data'] = daily_data['Dengue Data'] / 7

# Display the first few rows of the daily data
print(daily_data.head(20))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
daily_data.head()
```

```python
daily_data.shape
```

```python
daily_data = daily_data.dropna()
```

```python
daily_data.head()
```

```python
y = daily_data['Dengue Data']
x = daily_data[['Temperature', 'Humidity','Precipitation']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x ,y, test_size = 0.1)
```

```python
scale = StandardScaler()

X_train = scale.fit_transform(X_train)
X_test = scale.fit_transform(X_test)
```

##Linear Regression

```python
linear = LinearRegression()
```

```python
model = linear.fit(X_train, Y_train)
```

```python
model_pred = model.score(X_test, Y_test)
model_pred
```

##KNN

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=5)
```

```python
model_3 = knn.fit(X_train, Y_train)
```

```python
model_pred_3 = model_3.score(X_test, Y_test)
model_pred_3
```

##Random forest

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators=30)
```

```python
model_5 = rfr.fit(X_train, Y_train)
```

```python
model_pred_5 = model_5.score(X_test, Y_test)
model_pred_5
```
